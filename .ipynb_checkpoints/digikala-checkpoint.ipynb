```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from bidi.algorithm import get_display
from arabic_reshaper import reshape
```

```python
data = pd.read_csv('digi_kala.csv')
data
```

```python
data['DateTime_CartFinalize'] = pd.to_datetime(data['DateTime_CartFinalize'])
data['YearMonth'] = data['DateTime_CartFinalize'].map(lambda date: date.year*100 + date.month)
data
```

```python
items_per_month = data.groupby('YearMonth')['Quantity_item'].sum().reset_index()
items_per_month
```

```python
x = items_per_month['YearMonth'].to_numpy(dtype = str)
y = items_per_month['Quantity_item']

plt.figure(figsize=(30,10))
plt.plot(x , y, marker = 'o')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
customers = data.groupby('city_name_fa')['ID_Customer'].nunique().reset_index()
customers = customers.rename(columns={'ID_Customer' : 'Cus_Numbers'})
customers
```

```python
all_cus = customers['Cus_Numbers'].sum()
customers['Cus_Ratio'] = (customers['Cus_Numbers'] / all_cus) * 100

customers = customers.sort_values(['Cus_Ratio'], ascending=False).reset_index(drop= True)
customers
```

```python
others_city = pd.DataFrame({'city_name_fa' : 'others', 'Cus_Numbers' : customers['Cus_Numbers'][8:].sum(), 
                        'Cus_Ratio' : customers['Cus_Ratio'][8:].sum()}, index=[0])

others_city
```

```python
cities = pd.concat([customers[:8], others_city]).reset_index(drop= True)
cities
```

```python
labels = cities['city_name_fa']
#reshape the persians words
persian_labels = [get_display(reshape(label)) for label in labels]

plt.figure(figsize=(10, 10))
plt.pie(cities['Cus_Numbers'], labels= persian_labels, autopct= '%1.1f%%', textprops={'fontsize': 9})
plt.show()
```
